# OCRA script automation

```python
#this code block automatically runs the ORCA script
import os
import natsort
import subprocess

#for splitted_data testing purpose
directory="predicted_splitted_data/"
script1="../run1.sh"
script2="../extract_excited_energy.sh"

list_energies=[]
i=0
for file in natsort.natsorted(os.listdir(directory)):
    i+=1
    print(file)
    #changing the file name
    orca_config=open("test.xyz.inp","r+")
    config_data=orca_config.read()
    config_data=config_data.replace('benzene_0000.xyz',directory+file+"\n")
    orca_config.close()
    orca_config=open("test.xyz.inp","w").close()
    orca_config=open("test.xyz.inp","w")
    orca_config.write(config_data)
    orca_config.close()

    abc=subprocess.run(["sh", "run1.sh"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    # run the shell script and capture the output
    result = subprocess.run(["sh", "extract_excited_energy.sh"], stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    list_energies.append(result.stdout.decode('utf-8'))
    
    orca_config=open("test.xyz.inp","r+")
    config_data=orca_config.read()
    config_data=config_data.replace(directory+file,'benzene_0000.xyz')
    orca_config.close()
    orca_config=open("test.xyz.inp","w").close()
    orca_config=open("test.xyz.inp","w")
    orca_config.write(config_data)
    orca_config.close()
```
